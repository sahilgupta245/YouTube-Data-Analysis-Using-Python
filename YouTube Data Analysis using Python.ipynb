### Importing Libraries and Loading YouTube Dataset Using Pandas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("D:/Data Analyst/youtube dataset/IN_youtube_trending_data/IN_youtube_trending_data.csv")
df.head(3)
```

### Checking Datatypes and Performing Cleaning Process (By Removing Null Values & Droping Dublicate Values)

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
df["description"] = df["description"].fillna("Not Avaliable")
```

```python
df.duplicated().sum()
```

### Extracting publish_date, publish_time And trending_date From The Column publishedAt And trending_date.

```python
df["publish_date"] = df["publishedAt"].str.split("T").str[0]
df["publish_time"] = df["publishedAt"].str.split("T").str[1]
df["trendingDate"] = df["trending_date"].str.split("T").str[0]
df.head(5)
```

### Converting Their Datatypes From objects To datetime.

```python
df["publish_date"] = pd.to_datetime(df["publish_date"])
df["publish_time"] = pd.to_datetime(df["publish_date"])
df["trendingDate"] = pd.to_datetime(df["trendingDate"])
df.info()
```

## Find Top 6 Most Viewed YouTube Channels

### GroupingBy ChannelTitle And Aggrigate Sum Of view_count, Then Sorting Values In Descending Order.

```python
gb = df.groupby("channelTitle").agg({"view_count":"sum"})
gb = gb.sort_values(by = "view_count", ascending = False)
gb = gb.head(6)
gb
```

### Plotting Bar Chart

```python
channel_colors = {
    "T-Series": "#2D3740",          
    "MrBeast": "#E7943F",          
    "Sourav Joshi Vlogs": "#FE5C4F", 
    "HAR PAL GEO": "#635B4E",       
    "BLACKPINK": "#F7D290",
    "ARY Digital HD": "#B09471",
}

# Convert view count to billions/millions with labels
def format_labels(val):
    if val >= 1e9:
        return f"{val / 1e9:.2f}B"  # Billions
    elif val >= 1e6:
        return f"{val / 1e6:.2f}M"  # Millions
    return f'(val:.2f)'

gb["formatted_view_count"] = gb["view_count"].apply(format_labels)

plt.rcParams["figure.facecolor"] = "#FFFFFF"
plt.rcParams["text.color"] = "black"          
plt.rcParams["axes.labelcolor"] = "black"
plt.rcParams["xtick.color"] = "black"
plt.rcParams["ytick.color"] = "black"

plt.figure(figsize=(10, 7))
ax = sns.barplot(
    x="channelTitle",
    y="view_count",
    hue="channelTitle",
    data=gb,
    palette=[channel_colors[title] for title in gb.index],
    legend=False,
    edgecolor="black"
)

for i, bar in enumerate(ax.patches):
    height = bar.get_height()
    label = gb["formatted_view_count"].iloc[i]
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        height + 0.02 * height, 
        label,
        ha="center",
        fontsize=11,
        color="black",
        fontweight="bold",
    )

plt.title("Top 6 Most Viewed YouTube Channels", fontsize=16, fontweight="bold")
plt.ylabel("View Count (in Billions)", fontsize=9, fontweight="semibold")
plt.xlabel("Channel Title", fontsize=9, fontweight="semibold")
plt.xticks(fontsize=10, fontweight="bold")
plt.yticks(fontsize=10, fontweight="bold")

plt.tight_layout()
plt.show()
```

## Find Top 6 Most Disliked YouTube Channels

### GroupingBy ChannelTitle And Aggrigate Sum Of dislikes, Then Sorting Values In Descending Order.

```python
gb = df.groupby("channelTitle").agg({"dislikes": "sum"})
gb = gb.sort_values(by = "dislikes", ascending=False)
gb = gb.head(6)
gb
```

## Plotting Pie Chart

```python
channel_colors = {
    "FoxStarHindi": "#3C7BA4",          
    "T-Series": "#A6435C",        
    "Desi Music Factory": "#65ACDD", 
    "Zee Studios": "#A47E87",      
    "BLACKPINK": "#BBA037",
    "Zee Music Company": "#B0AD81", 
}

plt.figure(figsize=(9, 9))
plt.title("Top 6 Most Disliked YouTube Channels", fontsize=16, fontweight="bold")
plt.pie(
    gb["dislikes"],
    labels=gb.index,
    colors=[channel_colors[title] for title in gb.index],
    autopct="%1.2f%%",
    startangle=-150,
    textprops={'color': 'BLACK', 'fontweight':'bold'},
    wedgeprops={'edgecolor': 'black'},
)

plt.legend(
    labels=gb.index,  
    title="Channels",
    loc="upper left",  
    fontsize=10,  
    bbox_to_anchor=(1, 1)  
)

plt.tight_layout()
plt.show()
```
